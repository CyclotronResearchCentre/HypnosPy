```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
from heatmap import heatmap, corrplot
from scipy.stats import norm 
```

```python
data = pd.read_csv('hchs_quest_pa.csv')
#data = data.fillna(0).astype(int)
#data = data.astype(int)
data2 = pd.read_csv('hchs_quest_pa_one_hot.csv')
data = data.drop(columns=['Unnamed: 0', 'pid'])
data2 =data2.drop(columns=['Unnamed: 0', 'pid'])
```

```python
corrmat = data.corr() 
  
f, ax = plt.subplots(figsize =(14, 10),dpi = 350) 
sns.heatmap(corrmat, center= 0, ax = ax, cmap ="bwr", linewidths = 0.1) 
```
